# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.utils import resample
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# read in readmissions data provided
hospital_read_df = pd.read_csv('data/cms_hospital_readmissions.csv')
```

## Preliminary Analysis

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
clean_hospital_read_df = clean_hospital_read_df.sort_values('Number of Discharges')
```

```python
# generate a scatterplot for number of discharges vs. excess rate of readmissions
# lists work better with matplotlib scatterplot function
x = [a for a in clean_hospital_read_df['Number of Discharges'][81:-3]]
y = list(clean_hospital_read_df['Excess Readmission Ratio'][81:-3])

fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x, y,alpha=0.2)

ax.fill_between([0,350], 1.15, 2, facecolor='red', alpha = .15, interpolate=True)
ax.fill_between([800,2500], .5, .95, facecolor='green', alpha = .15, interpolate=True)

ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)

ax.grid(True)
fig.tight_layout()
```

### A

The preliminary analysis is good but does not have evidence to back up some of the claims that it makes, so I would add more evidence in the way of data visualization and hypothesis testing to build a more solid case for the final recommendations. A single plot is not enough to make a case, even for a simple preliminary report.  First, the data need to be cleaned.

```python
hospital_read_df.head()
```

```python
# original dataset
hospital_read_df.shape
```

```python
# keep rows that are not N/A for Excess Readmissions
hospital_read_df = hospital_read_df[hospital_read_df['Excess Readmission Ratio'].notna()]
```

```python
# keep rows that are not labeled 'Not Available' in Number of Discharges
hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
```

```python
# cleaned dataset
hospital_read_df.shape
```

```python
# Visualize Excess Readmission Ratio 
plt.style.use('seaborn')
hospital_read_df['Excess Readmission Ratio'].plot(kind='hist',bins=int(np.sqrt(len(hospital_read_df))))
plt.title('Excess Readmission Ratio')
plt.xlabel('Excess Readmission Ratio')
```

```python
# convert Number of Discharge to numeric
hospital_read_df['Number of Discharges'] = pd.to_numeric(hospital_read_df['Number of Discharges'])
hospital_read_df['Number of Discharges'].dtype
```

```python
hospital_read_df['Number of Discharges'].plot(kind='hist',bins=int(np.sqrt(len(hospital_read_df))))
plt.title('Histogram of Number of Discharges')
```

```python
sns.lmplot(x='Number of Discharges',y='Excess Readmission Ratio',data=hospital_read_df)
plt.title('Relationship between Number of Discharges and Excess Readmission Ratio')
```

There does appear to be a relationship between number of discharges and excess readmission ratio, as suggested by the preliminary report. However, we need to use spearman's r correlation instead of pearson beceause Number of Discharges is not normally distributed.

### B.1

The Hypothesis test is as follows

Significance level = 1%

**H0**: There is no correlation relationship between the two variables

**Ha**: There is some correlation relationship between the two variables

### B.2

```python
# perform hypothses test using spearman r test
corr, p = stats.spearmanr(hospital_read_df['Number of Discharges'],hospital_read_df['Excess Readmission Ratio'])
print('Spearman Correlation',corr)
print('P-Value:',p)
```

### B.3

At the 1% significance level, we reject the null hypothesis in favor of the alternative, there does appear to be a relationship between the number of discharges and the excess readmission ratio.

### B.4

As discharges increase, the Excess Readmission Ratio goes down, which is favorable for hospitals. This is probably due to the higher pool of resources that larger hospitals can draw from. Another factor is that larger hospitals may do better at attracting higher quality healthcare providers. With that in mind, what can the smaller hospitals do to reduce their readmissions? Taking our analysis into mind, they cannot simply up their # of discharges to solve this problem. Though there is a statistically significant effect, we would need to quantify the uncertainty around the actual mean value of correlation between the two variables to see if there is a practical significance here. There may not be a strong enough relationship practically to warrant the increase in cost from upping discharges. This could be quantified with a confidence interval and would allow policy makers to make recommendations around the metric. 

I would therefore not include the recommendation to consolidate hospitals without further study since we do not have an idea of the effect of that from a practical significance perspective. Making a consolidation recommendation could also make healthcare less accessible to people in rural or underserved areas, so more study would be needed before we could make such a recommendation without expecting negative results. Also, more data would be needed to identify the resources that could be given to smaller hospitals that would most effectively increase their ability to provide quality care, thereby reducing readmissions. The current dataset does not have that information available. At this point, providing training and more resources to smaller hospitals seems to be the most effective recommendation to make to reduce readmissions.

### B.5

The advantage of using a scatter plot is that we can easily see the negative trend in the data. However, it is very cluttered, and the highest density area of dots doesn't make an efficient use of our visual space. A seaborn jointplot with hexbins or a bivariate kde plot are better ways to visualize the data

```python
# joint plot
clean_plot = hospital_read_df[hospital_read_df['Number of Discharges'] < 2000]
sns.jointplot('Number of Discharges','Excess Readmission Ratio',data=clean_plot,kind='hex',ylim=(0.7,1.6))
```

```python
# kernel density plot
sns.kdeplot(hospital_read_df['Number of Discharges'],hospital_read_df['Excess Readmission Ratio'],shade=True)
plt.xlim(0,2000)
plt.ylim(0.7,1.6)
plt.title('KDE plot of Discharges vs Excess Readmission')
```
